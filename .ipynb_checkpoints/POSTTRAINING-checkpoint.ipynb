```python
train_data = pd.read_csv(HOME_DIR+'split_data/train_set.csv').values
valid_data = pd.read_csv(HOME_DIR+'split_data/val_set.csv').values
test_data = pd.read_csv(HOME_DIR+'split_data/test_set.csv').values
```

```python
class MainPyNet(nn.Module):
    
    def __init__(self, in_size, out_size):
        super(MainPyNet, self).__init__()
        
        self.fc1 = nn.Linear(in_size, num_neuro_1)
        self.fc2 = nn.Linear(num_neuro_1, num_neuro_2)
        self.fc3 = nn.Linear(num_neuro_2, num_neuro_3)
        self.fc4 = nn.Linear(num_neuro_3, out_size)
        
        #for m in self.modules():
          #  m.weight.data.normal_(0, 1)
          #  m.bias.data.fill_(1)
        
    def forward(self, x):
        x = F.tanh(self.fc1(x))
        x = F.tanh(self.fc2(x))
        x = F.tanh(self.fc3(x))
        x = F.tanh(self.fc4(x))
        return x
```

```python
# rand xavier
def init_weights(net, sigma, method='rand'):
    #print(sigma)
    for m in net.modules():
        if isinstance(m, nn.Linear):
            if method=='rand':
                m.weight.data.normal_(0, sigma)
            elif method=='xavier':
                m.weight.data.normal_(0, math.sqrt(2. / (m.weight.size(0)+m.weight.size(1))))
            else:
                break
                
    return net
```

```python
net = MainPyNet(IN_SIZE, OUT_SIZE)
X, y = Variable(torch.FloatTensor(train_data[:, :IN_SIZE])), Variable(torch.FloatTensor(train_data[:, IN_SIZE]))
```

```python
# 4.12a
sigma_list = [x/100 for x in range(10, 310, 10)]

param_list = {'mean': [], 'std': []}
for i in range(7):
    param_list['mean'].append([])
    param_list['std'].append([])
for s in sigma_list:
    net = init_weights(net, s, 'rand')
    
    out_list = []
    out_list.append(net.fc1(X)[:,3])
    out_list.append(net.fc1(X)[:,5])
    out_list.append(net.fc2(net.fc1(X))[:,0])
    out_list.append(net.fc2(net.fc1(X))[:,9])
    out_list.append(net.fc3(net.fc2(net.fc1(X)))[:,2])
    out_list.append(net.fc3(net.fc2(net.fc1(X)))[:,4])
    out_list.append(net.fc4(net.fc3(net.fc2(net.fc1(X)))))
    
    for i in range(len(out_list)):
        param_list['mean'][i].append(torch.mean(out_list[i].data))
        param_list['std'][i].append(torch.std(out_list[i].data))

        
xavier_std = {'sigma': [], 'mean':[], 'std': []}
for i in range(7):
    xavier_std['mean'].append([])
    xavier_std['std'].append([])
print('xavier:')
for m in net.modules():
    if isinstance(m, nn.Linear):
        xavier_std['sigma'].append(math.sqrt(2. / (m.weight.size(0)+m.weight.size(1))))
        print(m, ' - ', math.sqrt(2. / (m.weight.size(0)+m.weight.size(1))))
for s in xavier_std['sigma']:
    net = init_weights(net, s, 'rand')
    
    out_list = []
    out_list.append(net.fc1(X)[:,3])
    out_list.append(net.fc1(X)[:,5])
    out_list.append(net.fc2(net.fc1(X))[:,0])
    out_list.append(net.fc2(net.fc1(X))[:,9])
    out_list.append(net.fc3(net.fc2(net.fc1(X)))[:,2])
    out_list.append(net.fc3(net.fc2(net.fc1(X)))[:,4])
    out_list.append(net.fc4(net.fc3(net.fc2(net.fc1(X)))))
    
    for i in range(len(out_list)):
        xavier_std['mean'][i].append(torch.mean(out_list[i].data))
        xavier_std['std'][i].append(torch.std(out_list[i].data))
```

```python
import scipy
from scipy.optimize import linprog


def dual_task(module, X):
    outs=[]
    
    for i in range(len(X)):
        d=linprog(module.bias.data, -scipy.transpose(module.weight.data), X.data[i])
        print(d.x)
        outs.append(d.x)
        
    return outs
```
